```python
import numpy as np
import pandas as pd

from deap import base
from deap import creator
from deap import tools
```

```python
df_produtos = pd.DataFrame(tabela_produtos)
```

```python
def n_por_produto(min, max, elementos):
    return np.random.choice(range(min, max), size=elementos, replace=True).tolist()
```

```python
def avaliacao(individuo):
    individuo = individuo[0]
    total_proteinas = np.sum(df_produtos["gram_prot"].values * np.array([individuo]))
    total_carboidratos = np.sum(
        df_produtos["gram_carb"].values * np.array([individuo])
    )
    total_gorduras = np.sum(df_produtos["gram_gord"].values * np.array([individuo]))

    calorias = (
        caloria_proteina_grama * total_proteinas
        + caloria_carboidrato_grama * total_carboidratos
        + caloria_gordura_grama * total_gorduras
    )
    return abs(calorias - total_calorias), \
        abs(total_proteinas - grama_proteina), \
        abs(total_carboidratos - grama_carboidrato), \
        abs(total_gorduras - grama_gordura),
```

```python
# esses pessos são para minimização
# e são usados na função de avaliação
# preciso retornar uma tupla com os valores, um para cada objetivo/peso
pesos = (-1., -1. / 0.3, -1. / 0.2, -1./0.5)
creator.create("FitnessMin", base.Fitness, weights=pesos)
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# função de inicialização aleatória
toolbox.register(
    "n_por_produto",
    n_por_produto,
    df_produtos["min"].min(),
    df_produtos["max"].max() - 2,
    df_produtos.shape[0],
)

# cria um indivíduo com a função de inicialização
toolbox.register(
    "individual", tools.initRepeat, creator.Individual, toolbox.n_por_produto, n=1
)

# cria uma população com os indivíduos
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# função de avaliação
toolbox.register("evaluate", avaliacao)

# operador de cruzamento
toolbox.register("mate", tools.cxOnePoint)

# operador de mutação
toolbox.register(
    "mutate",
    tools.mutFlipBit,
    indpb=0.05,
)

# operador de seleção
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
def main():
    populacao = toolbox.population(n=300)

    # avaliando a população inteira
    fitness = list(map(toolbox.evaluate, populacao))

    # atribuindo o fitness para cada indivíduo
    for individuo, fit in zip(populacao, fitness):
        individuo.fitness.values = fit

    # CXPB  é a probabilidade de crossover
    # MUTPB é a probabilidade de mutação
    CXPB, MUTPB = 0.5, 0.2

    # extrai a lista de fitness
    fits = [individuo.fitness.values[0] for individuo in populacao]

    # geração
    geracao = 0

    # inicia o algoritmo
    while geracao < 5000:
        geracao += 1
        print(f"Geração {geracao}")

        # Seleciona a próxima geração de indivíduos
        filhos = toolbox.select(populacao, len(populacao))
        # Clona os indivíduos selecionados
        filhos = list(map(toolbox.clone, filhos))

        # aplicar crossover e mutação nos filhos
        for filho1, filho2 in zip(filhos[::2], filhos[1::2]):
            if np.random.rand() < CXPB:
                toolbox.mate(filho1[0], filho2[0])
                del filho1.fitness.values
                del filho2.fitness.values

        for mutante in filhos:
            if np.random.rand() < MUTPB:
                toolbox.mutate(mutante[0])
                del mutante.fitness.values

        # avalia os indivíduos com fitness inválido
        individuo_invalido = [filho for filho in filhos if not individuo.fitness.valid]
        fitnesses = map(toolbox.evaluate, individuo_invalido)
        for individuo, fit in zip(individuo_invalido, fitnesses):
            individuo.fitness.values = fit

        populacao[:] = filhos


        fits = [individuo.fitness.values[0] for individuo in populacao if individuo.fitness.valid]
        tamanho = len(populacao)
        media = sum(fits) / tamanho
        soma2 = sum(x * x for x in fits)
        std = abs(soma2 / tamanho - media ** 2) ** 0.5

    melhor = tools.selBest(populacao, 1)[0]
    return melhor
```

```python
melhor_solucao = main()
```

```python
df_produtos['resultado'] = pd.Series(melhor_solucao[0])
```
